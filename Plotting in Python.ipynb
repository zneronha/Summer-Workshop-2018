````python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

from palettable.colorbrewer.sequential import Oranges_9
````

````python
def factorial(x):
    if x<=1:
        return 1
    else: 
        return x * factorial (x-1)
    
````

````python
factorial(5)
````

````python
num_points = 50
xdata = np.zeros((num_points, 1))
ydata = np.zeros((num_points, 1))

# For Loop
for i in range(1, num_points):
    xdata[i] = i
    ydata[i] = factorial(i)
````

````python
fig = plt.figure(figsize=(4,3),dpi=200)

myriadpro = fm.FontProperties(fname='/usr/share/fonts/truetype/MyriadPro-Regular.ttf')

plt.semilogy(xdata,ydata,zorder = 1,alpha = 0.4,color=Oranges_9.mpl_colors[3])
plt.scatter(xdata,ydata,s=2,zorder = 2,color=Oranges_9.mpl_colors[5])
plt.title('Title',fontproperties = myriadpro)
plt.xlabel('X Axis',fontproperties = myriadpro)
plt.ylabel('Y Axis',fontproperties = myriadpro)

plt.tick_params(direction='out',length=5,width = 1.5)
plt.tight_layout()
plt.savefig('Plot.png')
````
